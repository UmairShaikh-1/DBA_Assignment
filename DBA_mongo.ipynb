### CRUD Operations Demonstration
This cell executes PyMongo operations directly using real values from the `orders`, `deliveries`, and `complaints` collections.

```python
# 1. CREATE: Insert a new order
new_order_doc = {
    "order_id": "O01251",
    "customer_id": "C0292",
    "service_type": "Passenger",
    "order_value": 150.75,
    "delay_minutes": 45.0,
    "delivery_status": "Pending"
}
db.orders.insert_one(new_order_doc)
print("CREATE: New order O01251 inserted.")
```

```python
# 2. READ: Find delayed orders (> 60 mins) and high severity complaints
delayed_orders = list(db.orders.find({"delay_minutes": {"$gt": 60}}).limit(2))
high_severity = list(db.complaints.find({"severity": "High"}).limit(2))

print(f"READ: Found {len(delayed_orders)} delayed orders and {len(high_severity)} high severity complaints.")

print("\n--- Delayed Orders Samples ---")
for order in delayed_orders:
    print(order)

print("\n--- High Severity Complaints Samples ---")
for complaint in high_severity:
    print(complaint)
```

```python
# 3. UPDATE: Change a complaint status
# Using complaint_id 'CP0001' from the dataset
# status will be updated from 'Open' to 'Resolved' and resolution days will change from '11' to '5'
db.complaints.update_one(
    {"complaint_id": "CP0001"},
    {"$set": {"status": "Resolved", "resolution_days": 5}}
)
```

```python
# 4. DELETE: Remove the test order created earlier
db.orders.delete_one({"order_id": "O01251"})
print("DELETE: Test order O01251 removed.")
```

### Impact of CRUD Operations on Operational Workflows

CRUD (Create, Read, Update, Delete) operations are the fundementals of data management in MongoDB. In a logistics and service environment like NorthStar:

*   **Create**: Allows for real-time capturing of new customer orders and complaints as they occur.
*   **Read**: Enables operational teams to filter and identify critical issues, such as finding high-severity complaints or identifying significantly delayed deliveries for immediate intervention.
*   **Update**: Supports dynamic workflows, such as moving a complaint from 'Open' to 'Resolved' or appending incident logs to an ongoing delivery without overwriting existing data.
*   **Delete**: Helps to ensure all contradictory and unwanted records are removed, maintaining a clean and accurate production environment.

## Query Optimization with Indexing
In MongoDB, indexing is a critical technique to ensure queries run efficiently by reducing the number of documents scanned.

```python
# 1. Slow Query Before Optimization (Scanning all documents)
print("--- Before Index: finding 'Delayed' orders ---")
explain_before = db.orders.find({"delivery_status": "Delayed"}).explain()
print(f"Winning Plan: {explain_before['queryPlanner']['winningPlan']['stage']}")
print(f"Documents Scanned: {explain_before['executionStats']['totalDocsExamined']}")

# 2. Create Index for delivery status
db.orders.create_index([("delivery_status", 1)])
print("\nIndex created on 'delivery_status'.")

# 3. Query After Optimization (creating index)
print("\n--- After Index: finding 'Delayed' orders ---")
explain_after = db.orders.find({"delivery_status": "Delayed"}).explain()
print(f"Winning Plan: {explain_after['queryPlanner']['winningPlan']['stage']}")
print(f"Documents Scanned: {explain_after['executionStats']['totalDocsExamined']}")
```

```python
# Second Example: Querying by order_id
print("--- Before Index: query by order_id ---")
explain_order_before = db.orders.find({"order_id": "O00001"}).explain()
print(f"Winning Plan: {explain_order_before['queryPlanner']['winningPlan']['stage']}")

# Create Index for order id
db.orders.create_index([("order_id", 1)])
print("\nIndex created on 'order_id'.")

# Using index
print("\n--- After Index: query by order_id ---")
explain_order_after = db.orders.find({"order_id": "O00001"}).explain()
print(f"Winning Plan: {explain_order_after['queryPlanner']['winningPlan']['stage']}")
print(f"Index Used: {'IXSCAN' in str(explain_order_after['queryPlanner']['winningPlan'])}")
```

### How Indexing Improves Performance

Indexing in MongoDB significantly boosts performance by creating a small data structure that stores a subset of the collection's data.

*   **Reduction in Collection Scans**: Without an index, MongoDB must perform a 'COLLSCAN' (Collection Scan), reading every document in a collection to find matches. With an index, it performs an 'IXSCAN' (Index Scan), jumping directly to the relevant data.
*   **Lower Resource Consumption**: By scanning fewer documents, the database uses less CPU and Memory, allowing it to run faster and handle more requests.
*   **Winning Plan**: The `explain()` method reveals the 'winning plan' (best query plan) chosen by the MongoDB optimizer. Transitioning from `COLLSCAN` to `IXSCAN` is the primary goal of query optimization, ensuring the system remains scalable as the dataset grows.
